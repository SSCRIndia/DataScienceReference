```python
startdate = datetime.datetime.strptime('2018-12-29', '%Y-%m-%d').date()
enddate = datetime.datetime.strptime('2021-01-31', '%Y-%m-%d').date()
print(startdate, enddate)
```

```python
from dateutil.relativedelta import relativedelta
```

```python
dataset = pd.DataFrame()
while startdate<=enddate:
    file_name = 'http://web.mta.info/developers/data/nyct/turnstile/turnstile_'+ startdate.strftime('%y%m%d') +'.txt'
    print(file_name)
    tdata = pd.read_csv(file_name)
    tdata.columns = ['CA', 'UNIT', 'SCP', 'STATION', 'LINENAME', 'DIVISION', 'DATE', 'TIME', 'DESC', 'ENTRIES', 'EXITS']
    tdata['file_name'] = startdate
    tdata.to_csv(str(startdate)+'_data.csv', sep = '|', index = False)
    startdate = startdate+relativedelta(days = 7)
    
```
